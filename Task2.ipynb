## 1. **Importing libraries**

```python
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

## 2. **Load Dataset**

     [Here loading Iris dataset]

```python
data=pd.read_csv('Iris.csv')
data.drop(['Id'],axis=1,inplace=True)
```

3. **Reading the dataset**

    [Displaying the first 5 rows of  dataset.]

```python
data.drop_duplicates(inplace=True)   #removing duplicates from the data frame
```

## 4. **Visualizing** 

```python
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data,hue='Species')     #To plot multiple pairwise bivariate distributions in a dataset we are using pairplot
```

```python
data.corr()     #used to find the pairwise correlation of all columns in the dataframe
```

```python
data=data.iloc[:,[0,1,2,3]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data)
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
    
wcss
```

## 5. **Plotting The Elbow Graph To Find Optimal Value Of 'K**'

```python
plt.figure(figsize=(13,6))
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('The Elbow Method for Optimal  K')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

------->The optimum value for K would be 3.



## 6. **Implementing K-Means Clustering**
       Applying k-means to the dataset. Creating the k-means classifier

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(data)
y_kmeans
```

```python
fig = plt.figure(figsize=(13, 7))    #Plotting graph
plt.title('Clusters with Centroids',fontweight ='bold', fontsize=20)
plt.scatter(data[y_kmeans == 0, 2], data[y_kmeans == 0, 3], s = 100, c = 'cyan', label = 'Iris-versicolour')
plt.scatter(data[y_kmeans == 1, 2], data[y_kmeans == 1, 3], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(data[y_kmeans == 2, 2], data[y_kmeans == 2, 3],s = 100, c = 'orange', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:,3], s = 300, c = 'maroon',marker='*',label = 'Centroids')
plt.title('K-Means Clustering using n_cluster=3')
plt.ylabel('Petal Length in cm')
plt.xlabel('SepalLength in cm')
plt.legend()
```
